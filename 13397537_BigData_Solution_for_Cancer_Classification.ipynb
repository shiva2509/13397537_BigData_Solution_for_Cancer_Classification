```
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Cancer Classification using Big Data") \
    .getOrCreate()
```

```
# File location and type
file_location = "/FileStore/tables/file1-1.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

from pyspark.sql.functions import col

# Load the entire dataset
data = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

number_of_features = 100  

# Select the 'C' column and the specified number of 'g' columns
selected_columns = ['Class'] + [f'gene_{i}' for i in range(number_of_features)]
df = data.select(selected_columns)


# Display the selected DataFrame
display(df.limit(5))
```

```
df.printSchema()
```

```
from pyspark.sql.functions import isnan, when, count, col

# Counting the missing values in each column
missing_values = df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns])

# Checking if any column has missing values
missing_values_present = any(row[c] > 0 for row in missing_values.collect() for c in missing_values.columns)

# Print the appropriate message
if missing_values_present:
    print("Missing Values")
else:
    print("No Missing Values")

```

```
from pyspark.sql.functions import isnan, when, count, col

# Fill missing values with 0
df_filled = df.fillna(0)

# Counting the missing values in each column after filling
missing_values_after_filling = df_filled.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df_filled.columns])

# Checking if any column still has missing values
missing_values_still_present = any(row[c] > 0 for row in missing_values_after_filling.collect() for c in missing_values_after_filling.columns)

# Print the appropriate message
if missing_values_still_present:
    print("Missing Values")
else:
    print("No Missing Values")

```

```
from pyspark.ml.classification import RandomForestClassifier, LogisticRegression, DecisionTreeClassifier
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.sql import functions as F
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_pca is your DataFrame with PCA features and label 'C'

# Indexing labels
stringIndexer = StringIndexer(inputCol="Class", outputCol="indexedLabel")
si_model = stringIndexer.fit(df_pca)
data1 = si_model.transform(df_pca)

# Split data
(train_data, test_data) = data1.randomSplit([0.8, 0.2])

# Initialize classifiers
rf = RandomForestClassifier(labelCol="indexedLabel", featuresCol="pcaFeatures")
lr = LogisticRegression(labelCol="indexedLabel", featuresCol="pcaFeatures")
dt = DecisionTreeClassifier(labelCol="indexedLabel", featuresCol="pcaFeatures")

# List of models
models = [rf, lr, dt]
metrics = []
confusion_matrices = {}


import time
metrics = []
confusion_matrices = {}
model_training_times = {}  # Dictionary to store training times

# Train and evaluate each model
for model in models:
    # Train model

    start_time = time.time()  # Start time
    # Train model
    model_pipeline = model.fit(train_data)
    end_time = time.time() 
    training_time = end_time - start_time  # Calculate training duration
    model_training_times[model.__class__.__name__] = training_time   

    # Make predictions
    predictions = model_pipeline.transform(test_data)

    # Evaluate predictions
    evaluator = MulticlassClassificationEvaluator(labelCol="indexedLabel", predictionCol="prediction")
    accuracy = evaluator.evaluate(predictions, {evaluator.metricName: "accuracy"})
    precision = evaluator.evaluate(predictions, {evaluator.metricName: "weightedPrecision"})
    recall = evaluator.evaluate(predictions, {evaluator.metricName: "weightedRecall"})
    f1 = evaluator.evaluate(predictions, {evaluator.metricName: "f1"})
    metrics.append((model.__class__.__name__, accuracy, precision, recall, f1))

    # Confusion Matrix
    predictionAndLabels = predictions.select("prediction", "indexedLabel").rdd
    metrics_mllib = MulticlassMetrics(predictionAndLabels)
    confusion_matrix = metrics_mllib.confusionMatrix().toArray()
    confusion_matrices[model.__class__.__name__] = confusion_matrix

# Display metrics
results_df = pd.DataFrame(metrics, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
sorted_results = results_df.sort_values(by='Accuracy', ascending=False)
print(sorted_results)

for model_name, training_time in model_training_times.items():
    print(f"{model_name}: Training Time = {training_time:.2f} seconds")


# Plot Confusion Matrices
for model_name, cm in confusion_matrices.items():
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='g')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```
